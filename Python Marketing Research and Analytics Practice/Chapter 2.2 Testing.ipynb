```python
#making frames
a = [7, 4, 2, 22, -12]
b = [34, 2, 9, -5, -18]
ab_df = pd.DataFrame({'a': a, 'b': b})
ab_df
ab_df = pd.DataFrame({'a': a, 'b': b},
index=pd.date_range(start=start_time,
normalize=True, periods=5))
ab_df
```
