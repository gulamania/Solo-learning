File saving with Pickles, and CSVs. apparently pickle is a standard format for saving crap that I've never heard of. This is what I get for self teaching through stack overflow.

```python
import pickle
#saving a dataset to a pickle. what constitutes a salty pickle file?
store_df = pd.read_csv('http://bit.ly/PMR-ch2')
f = open('store_df.pkl', 'wb')
pickle.dump(store_df, f)
f.close()
```

```python
#opening a saved file with pickle
with open('store_df.pkl', 'rb') as f:
    store_df_reload = pickle.load(f)
store_df_reload
#nice 
#for more direct use the following exist:
#   store_df.to_pickle('filepath')
#   store_df_reload = pd.read_pickle('filepath')
```

```python
#csv
store_df.to_csv('store_df.csv', index = False)
#use store_df.to_csv() to send to terminal
```

```python
#reading csvs
store_df_from_csv = pd.read_csv('store_df.csv')
store_df_from_csv
#note that csvs dont save column type information.
#you may have to change column type manually later bc of this
```
